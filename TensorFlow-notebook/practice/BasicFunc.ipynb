```python
# 创建全0 全1的向量
h1: Tensor = tf.zeros([2, 2])
h2: Tensor = tf.ones([2, 2])
print(f"h1: {h1}")
print(f"h2: {h2}")

print("=======================")

h3: Tensor = tf.zeros_like(h2)
h4: Tensor = tf.ones_like(h1)
print(f"h3: {h3}")
print(f"h4: {h4}")

print("=======================")

# 创建自定义数值的张量
h5: Tensor = tf.fill([1], -1)   # 第一个参数是维度, 第二个参数是数值
h6: Tensor = tf.fill([2,2], 99)
print(f"h5: {h5}")
print(f"h6: {h6}")
```

```python
# ragged tesnor, 用于创建不规则的向量
r = tf.ragged.constant([[11, 12], [21, 22, 23], [], [4]])
print(r)
print(r[1])
print(r[1: 2])

# ops on ragged tensor
r2 = tf.ragged.constant([[11, 12], [21, 22, 23], []])
print(tf.concat([r, r2], axis=0))       

# ragged转化为tensor时会填充0
print(r.to_tensor())
```

```python
# sparse tensor

s = tf.SparseTensor(indices=[[0, 1], [1, 0], [2 ,3]],
                   values=[1., 2., 3.], 
                   dense_shape=[3, 4])
print(s)
# 转化为密集矩阵
# 注意indices的必须是排好序的，才能进行to_dense操作
print(tf.sparse.to_dense(s))

s2 = s * 2.
print(s2, "\n")

s4 = tf.constant([[10., 20.],
                  [30., 40.],
                  [50., 60.],
                  [70., 80.]])

print(tf.sparse.sparse_dense_matmul(s, s4))
```

```python
# 简单的前向传播算法计算流程
# 模拟MNIST数据集 训练三层神经网络 输入节点为784, 第一层输出节点256, 第二层输出节点数128, 第三层输出10
w1: Variable = tf.Variable(tf.random.truncated_normal([784, 256], stddev=0.1))
b1: Variable = tf.Variable(tf.zeros([256]))
w2: Variable = tf.Variable(tf.random.truncated_normal([256, 128], stddev=0.1))
b2: Variable = tf.Variable(tf.zeros([128]))
w3: Variable = tf.Variable(tf.random.truncated_normal([128, 10], stddev=0.1))
b3: Variable = tf.Variable(tf.zeros([10]))
    
# 先转换输入值的维度
x: Tensor = tf.random.normal([30, 28, 28])
x = tf.reshape(x, [30, -1])

h1 = x @ w1 + tf.broadcast_to(b1, [30, 256])
h1 = tf.nn.relu(h1)

h2 = h1 @ w2 + tf.broadcast_to(b2, [h1.shape[0], 128])
h2 = tf.nn.relu(h2)

h3 = h2 @ w3 + tf.broadcast_to(b3, [h2.shape[0], 10])
out = tf.nn.relu(h3)
```
