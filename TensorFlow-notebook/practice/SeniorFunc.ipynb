```python
import numpy as np
import tensorflow as tf

from tensorflow import Tensor, Variable
from typing import Dict, List
```

```python
# 合并(合并操作可以在任意维度上进行, 唯一的约束是非合并维度的长度必须一致/ 不会创建新维度)
a: Tensor = tf.random.normal([4, 35, 8])
b: Tensor = tf.random.normal([6, 35, 8])
    
tf.concat([a, b], axis=0)
```

```python
# 合并时,需要创建新维度用tf.stack
a: Tensor = tf.random.normal([35, 4])
b: Tensor = tf.random.normal([35, 4])
    
tf.stack([a, b], axis=0)
```

```python
# 分割
x: Tensor = tf.random.normal([10, 35, 8])

# 等长切割
z1: List[Tensor] = tf.split(x, axis=0, num_or_size_splits=10)
z2: List[Tensor] = tf.split(x, axis=1, num_or_size_splits=[30, 5])
```

```python
# 如果希望在某个维度上全部按长度为1的方式分割, 可直接用tf.unstack(会降一个维度)
x: Tensor = tf.random.normal([10, 35, 8])
tf.unstack(x, axis=0)
```

```python
# 范数
x: Tensor = tf.ones([2, 2])
print(x)
    
# 1-范数
tf.norm(x, ord=1)
print(f"1-范数: {tf.norm(x, ord=1)}")

# 2-范数
tf.norm(x, ord=2)
print(f"2-范数: {tf.norm(x, ord=2)}")

# ∞-范数
tf.norm(x, ord=np.inf)
print(f"∞-范数: {tf.norm(x, ord=np.inf)}")
```

```python
# 求某个维度上最大最小、均值、和
x: Tensor = tf.random.normal([4, 10, 3])
    
tf.reduce_max(x, axis=0)
print(f"0维度的最大值: {tf.reduce_max(x, axis=0)}")

tf.reduce_min(x, axis=0)

tf.reduce_mean(x, axis=0)

tf.reduce_sum(x, axis=2)
```

```python
out: Tensor = tf.random.normal([4, 10])  # 预测输出
print(f"out: \n{out}")
y: Tensor = tf.constant([1, 2, 2, 0])    # 真实标签
y = tf.one_hot(y, depth=10)
print(f"经过独热编码后的真实标签: \n{y}")

print(f"validate: {tf.norm((out[0]-y[0]), ord=2)**2 /10}")

loss = tf.keras.losses.mse(y, out)
loss
```

#### tf.nn.sotfmax 是一种归一化方式

```python
# 获取最值所在的索引号
out: Tensor = tf.random.normal([2, 10])
print(f"out: \n{out}")

out = tf.nn.softmax(out, axis=1)
print(f"out after softmax: \n{out}")

tf.argmax(out, axis=1)
```

```python
# 张量比较
out: Tensor = tf.random.normal([100, 10])
out = tf.nn.softmax(out, axis=1)
pred: Tensor = tf.argmax(out, axis=1)

y: Tensor = tf.random.uniform([100], dtype=tf.int64, maxval=10)
print(f"y: \n{y}")

z: Tensor = tf.equal(pred, y)
print(f"bool: {z}")

# 统计True的个数(先将bool转换成int, 然后利用sum来统计True的个数)
tf.reduce_sum(tf.cast(z, dtype=tf.int16))
 
```

```python
# 填充

a: Tensor = tf.constant([1, 2, 3, 4, 5, 6])
b: Tensor = tf.constant([7, 8, 1, 6])

b = tf.pad(b, paddings=[[0, 2]])

tf.stack([a, b], axis=0)
```

```python
# 数据限幅

a: Tensor = tf.range(9)
print(f"a: \n{a}")    

tf.maximum(a, 5)        # 下限幅为5
print(f"tf.maximum: \n{tf.maximum(a, 5)}")

tf.minimum(a, 5)        # 上限幅为5
print(f"tf.minimum: \n{tf.minimum(a, 5)}")

tf.clip_by_value(a, 2, 7)
```

```python
# 假设有4个班级, 每个班有35个学生, 8门科目
x = tf.random.uniform([4, 35, 8], maxval=100, dtype=tf.int32)
print(f"x: \n{x}")

# tf.gather适合没有规则的场景
# 查看第1、2班级的成绩
x1 = tf.gather(x, [1, 2], axis=0)
print(f"x1: \n{x1}")

# tf.gather_nd多维度收集数据
x2 = tf.gather_nd(x, [[1, 1], [2, 2], [3, 3]])
print(f"x2: \n{x2}")
```

```python
#tf.boolean_mask
tf.boolean_mask(x, mask=[True, False, True, False], axis=0)
```

```python
a: Tensor = tf.ones([3, 3])
b: Tensor = tf.zeros([3, 3])

cond: tf.constant = tf.constant([[True, False, False], [False, False, True], [False, True, True]])
x1: Tensor = tf.where(cond, a, b)
print(f"x1: \n{x1}")

# 若a、b为None时, tf.where will return index of true in cond
x2: Tensor = tf.where(cond)
print(f"x2: \n{x2}")
```
